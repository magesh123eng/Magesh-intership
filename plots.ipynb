```python
import pandas as pd 
df=pd.read_csv("/Users/apple/Documents/MAGI 2.csv")
print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(data=df,x="AGE",y="MOBILE")
plt.title("scatter plot")
plt.show()
```

```python
sns.lineplot(data=df,x="AGE",y="MOBILE")
plt.title("lineplot")
```

```python
sns.boxplot(data=df,x="AGE",y="MOBILE")
plt.title("boxplot")
```

```python
sns.histplot(data=df,x="AGE",y="MOBILE")
plt.title("histplot")
```

```python
sns.kdeplot(data=df,x="AGE",y="MOBILE")
plt.title("kdeplot")
```
