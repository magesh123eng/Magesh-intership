```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[5,6,7,8,9]
plt.plot(x,y)
plt.title("simple code")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
plt.pie(x)
plt.title("simple code")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
