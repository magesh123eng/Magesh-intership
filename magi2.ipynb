```python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset("tips")
sns.countplot(x="day", data=tips)
plt.title("Count Plot of Days")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset("tips")
sns.stripplot(x="day", data=tips)
plt.title("Count Plot of Days")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset("tips")
sns.boxplot(x="day", data=tips)
plt.title("Count Plot of Days")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset("tips")
sns.swarmplot(x="day", data=tips)
plt.title("Count Plot of Days")
plt.show()
```
